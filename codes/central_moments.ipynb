Central moments are crucial for analysing and interpreting of the probability density. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moment as mt
import seaborn as sns
```

```python
data_n = np.random.randn(10000)
r = np.random.rand(10) + 2
e = np.random.rand(100) + 1
d = np.random.rand(300)
dd = np.append(d, e)
data_s = np.append(dd,r)

plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.histplot(data_n , stat = 'density', kde = 1)

plt.subplot(1,2,2)
sns.histplot(data_s , stat = 'density', kde = 1)
```

The first central momentum is always zero:

```python
mn_1 = mt(data_n, moment = 1)
ms_1 = mt(data_s , moment = 1)
print (mn_1)
print (ms_1)
```

The second central momentum measures variance:

```python
mn_2 = mt(data_n, moment = 2) # notice that according to numpy documentation, the varicance of this numbers is 1
ms_2 = mt(data_s , moment = 2)

mnn_2 = np.var(data_n)
mss_2 = np.var(data_s)

print (mn_2)
print (ms_2)

print (mnn_2)
print (mss_2) # you can infer that second momentum is really the variance!
```

The third central momentum measures skewness or the asymmetry of the pdf. In my example, normal distribution is approximately symmetric so it should be zero (or a tiny negative zero. This is because the left tail of graph is a bit longer than the right tail). But for the other sample, it should be positive (since obviously the tail is on the right side):

```python
mn_3 = mt(data_n, moment = 3) 
ms_3 = mt(data_s , moment = 3)

print (mn_3)
print (ms_3)
```

The fourth central momentum measures the tailedness of the pfd. So, the sharper peaks a pdf has, the greater Kurtosis it will have:

```python
mn_4 = mt(data_n, moment = 4) 
ms_4 = mt(data_s , moment = 4)

print (mn_4)
print (ms_4)
```

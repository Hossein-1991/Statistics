```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

The central limit theorem (CLT) is one of the most important results in probability theory. It states that, under certain conditions (finite mean and variance), the sum of a large number of random variables is approximately normal

```python
main_data = np.random.uniform(0,100,1000000) # our main data is a uniform distribution
c = []
d = np.zeros(50)
n = [2,25,100]
w = 0
fig , axes = plt.subplots(1,3,figsize=(20,8))
for t in n:
  for i in range(0,t):
    sampled_data = np.random.choice(main_data, size = 50 , replace = False)
    c.append(sampled_data)
  c = np.array(c)
  for j in range(0,t):
    d += c[j]
  sns.histplot(data = d , kde = 1, ax = axes[w])
  axes[w].set_title("number of samples {}".format(t))
  w += 1
  c = []
  d = np.zeros(50)
plt.show()
```

As you can see, the total distribution of variables is normal if you increase the number of samples. 

But this is true even if your samples are not from the same distribution:

```python
a = np.random.uniform(0,100,10000)
b = np.random.randint(0,100,10000)
c = np.random.beta(1,2,10000)
cc = []
d = np.zeros(50)
n = [2,10,100]
w = 0
fig , axes = plt.subplots(1,3,figsize=(20,8))

for t in n:
  for i in range(0,t):
    r1 = np.random.choice(a, size = 50 , replace = False)
    r2 = np.random.choice(b, size = 50 , replace = False)
    r3 = np.random.choice(c, size = 50 , replace = False)
    r = r1 + r2 + r3
    cc.append(r)
  cc = np.array(cc)
  for j in range(0,t):
    d += cc[j]
  sns.histplot(data = d , kde = 1, ax = axes[w])
  axes[w].set_title("number of samples {}".format(t))
  w += 1
  cc = []
  d = np.zeros(50)
plt.show()
```

The mentioned pattern is obvious here too.

But there is another wonderful fact as well! If your main_data stay unchanged, your average mean will get closer to the main_data mean as you increase the number of samples:

```python
main_data = np.random.randint(0,40,100000)
print ('mean of the main_data: ', np.mean(main_data))

c = []
d = np.zeros(20)
n = [2,15,50]

for t in n:
  for i in range(0,t):
    sampled_data = np.random.choice(main_data, size = 20 , replace = False)
    c.append(sampled_data)
  c = np.array(c)
  for j in range(0,t):
    d += c[j]
  print ('number of samples: {}/ mean: {}'.format(t,np.mean(d)/t))
  c = []
  d = np.zeros(20)
```

A power spectrum displays the coefficients for each frequency measured by the FFT as a graph of power values (From: Supplements to Clinical Neurophysiology, 2013)

For 1D cases, it is kind of the same as fourier transform and does not have additional informtation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import rfft
from scipy.fft import rfftfreq
```

```python
x = np.arange(0,20,0.1)
sinn = np.sin(2*np.pi*x) # frequency = 1

f_sinn = rfft(sinn) # rfft stands for real fast fourier transform
f_k = rfftfreq(len(sinn), 20/len(sinn))

power_coeff = (np.abs(f_sinn))**2

plt.plot(f_k, power_coeff)
```

```python
# importing data:
data = pd.read_csv('/content/drive/MyDrive/Sunspots.csv')
```

```python
data = data.Monthly.values
```

```python
data = (data-np.mean(data))/np.std(data)
```

```python
# implementing fourier transform
f_data = rfft(data)
f_k = rfftfreq(len(data), 1)
```

```python
# calculating power spectrum coefficient:
power_coeff = (np.abs(f_data))**2
plt.plot(f_k, power_coeff)
plt.xlim(0,0.1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Some expressions!:

pdf : When the outcomes of a random variable are continuous 

cdf : The probability that a random varibale takes a value less or equal to some number 


```python
dataa = np.random.randn(10000) # generating data

plt.figure(figsize=(19,6))
plt.subplot(1,3,1)
sns.histplot(dataa, stat = 'count' , binwidth = 0.2) # count: show the number of observations in each bin
plt.title('distribution frequency (df)')

plt.subplot(1,3,2)
sns.histplot(dataa , stat = 'density' , binwidth = 0.2) # density: normalize such that the total area of the histogram equals 1
plt.title('probability density function (pdf)')

plt.subplot(1,3,3)
sns.histplot(dataa, stat = 'density' , binwidth = 0.2, cumulative=1)
plt.title('cumulative distribution function (CDF)')
```
